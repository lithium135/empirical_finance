```python
import xlsxwriter
import pandas as pd
import math
import time
import shutil
from tqdm import tqdm

from openpyxl import load_workbook
from win32com.client import DispatchEx, Dispatch
from win32api import LoadLibrary
from IPython.display import HTML, Javascript, display

PATH = "C:/Users/jackl/OneDrive/Documents/finance_research/japan_qe/"
```

```python
def display_file(p):
    wb = load_workbook(p, data_only=True)
    sh = wb["Sheet1"]
    return pd.DataFrame(sh.values)
```

```python
nk_df = pd.read_csv(PATH+'raw_data/capital_iq/nk_df.csv')
```

```python
ticker_list = list(nk_df['ticker'])
```

```python
ticker_list = [1332, 1333]
month_list = ['06']
```

## Action Items

```python
for ticker in tqdm(ticker_list):
    for month in month_list:
        p = PATH+'raw_data/capital_iq/ownership_files/' + str(ticker) + "_" + str(month) + '.xlsx'

        excel = DispatchEx('Excel.Application')
        excel.Visible = True

        workbook = excel.Workbooks.Open(p)

        time.sleep(10)
        excel.CalculateFull()

        time.sleep(10)
        workbook.Save()

        workbook.Close()
        excel.Quit()
```
